ChatGPT Prompt: beri kode untuk membuat dataframe dari file csv dengan nama file gender.csv

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset1/gender.csv')

# Menampilkan lima baris pertama dataframe
print(df.head())
```

ChatGPT Prompt: beri kode untuk menampilkan feature statistic

```python
# Menampilkan feature statistic dari fitur-fitur dalam dataframe
print(df.describe())
```

```python
df.info()
```

ChatGPT Prompt: Beri kode untuk menampilkan boxplot dari fitur-fitur numerik

```python
# Mengambil hanya fitur numerik
numeric_features = df.select_dtypes(include=['int64', 'float64'])

# Menampilkan boxplot untuk setiap fitur numerik
for column in numeric_features.columns:
    plt.figure(figsize=(8, 6))
    df.boxplot(column=[column])
    plt.title(f'Boxplot dari {column}')
    plt.ylabel(column)
    plt.show()
```

ChatGPT Prompt: beri kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Mengatur gaya plot seaborn
sns.set(style="ticks", palette='Reds')
# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(numeric_features)
plt.suptitle('Pairplot of Numeric Features', y=1.02, fontsize=16)
plt.show()
```

hatGPT Prompt: Beri kode untuk menampilkan fitur numerik dalam bentuk distribution plot

```python
# Menampilkan distribution plot untuk setiap fitur numerik
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Distribution Plot untuk Fitur Numerik', fontsize=16)
plt.show()
```

ChatGPT Prompt: beri kode untuk menampilkan violin plot dari fitu-fitur numerik

```python
# Menampilkan violin plot dari fitur-fitur numerik
plt.figure(figsize=(14, 10))
sns.set(style="whitegrid")
sns.violinplot(data=df.select_dtypes(include=['float64', 'int64']), inner="points", linewidth=1.5)
plt.title('Violin Plot dari Fitur-Fitur Numerik')
plt.xlabel('Fitur')
plt.ylabel('Nilai')
plt.show()
```

ChatGPT Prompt: beri kode untuk menampilkan heat map dari fitu-fitur numerik

```python
# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Menampilkan heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap dari Korelasi Fitur-Fitur Numerik')
plt.show()
```
