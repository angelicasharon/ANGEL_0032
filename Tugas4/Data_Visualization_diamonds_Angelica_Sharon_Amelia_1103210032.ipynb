ChatGPT Prompt: Beri kode untuk menampilkan boxplot dari fitur-fitur numerik

```python
# Menampilkan boxplot untuk setiap fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.set_palette("Blues")
    sns.boxplot(x=df[feature])
    plt.title('Boxplot of Numeric Features ' + feature)
    plt.show()
```

ChatGPT Prompt: beri kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Mengatur gaya plot seaborn
sns.set(style="ticks", palette='Reds')
# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(numeric_features)
plt.suptitle('Pairplot of Numeric Features', y=1.02, fontsize=16)
plt.show()
```

ChatGPT Prompt: Beri kode untuk menampilkan fitur numerik dalam bentuk distribution plot

```python
# Menampilkan distribution plot untuk setiap fitur numerik
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Distribution Plot untuk Fitur Numerik', fontsize=16)
plt.show()
```

ChatGPT Prompt: beri kode untuk menampilkan violin plot dari fitu-fitur numerik

```python
# Menampilkan violin plot untuk semua fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=df[feature], inner='quartile')
    plt.title(f'Violin Plot untuk Fitur Numerik {feature}')
    plt.xlabel(feature)
    plt.show()
```
